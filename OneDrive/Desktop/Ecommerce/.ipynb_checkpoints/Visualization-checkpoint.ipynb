```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cleaned_ecommerce_data.csv")
```

```python
platform_revenue = df.groupby('Platform')['Order Value (INR)'].sum()
monthly_sales = df.groupby('Month')['Order Value (INR)'].sum()
```

```python
print(platform_revenue)
```

```python
refund_counts = df['Refund Requested'].value_counts()

print("\nRefund Requested Counts:")
print()
```

```python
delay_counts = df['Delivery Delay'].value_counts()

print("Delivery Delay Counts:")
print(delay_counts)
```

```python
comparison = df.groupby('Platform').agg({
    'Order Value (INR)': ['mean', 'count'],
    'Delivery Delay': delay_counts,
    'Refund Requested': 
})
```

```python
plt.figure()
comparison.plot(kind='bar')
plt.title("Revenue by Platform")
plt.xlabel("Platform")
plt.ylabel("Revenue")
plt.show()
```

# Revenue by platform

```python
plt.figure()
platform_revenue.plot(kind='bar')
plt.title("Revenue by Platform")
plt.xlabel("Platform")
plt.ylabel("Revenue")
plt.show()
```

```python
average = df.groupby('Platform')['Order Value (INR)'].mean()
```

```python
plt.figure()
average.plot(kind='bar')
plt.title("Revenue by Platform")
plt.xlabel("Platform")
plt.ylabel("Revenue")
plt.show()
```

```python
ax = platform_revenue.plot(kind='bar')

for i, v in enumerate(platform_revenue):
    ax.text(i, v, str(round(v, 0)), ha='center')

plt.show()
```

```python
(platform_revenue / platform_revenue.sum() * 100).plot(kind='bar')
plt.title("Revenue Share (%)")
plt.show()
```

```python
platform_revenue.sort_values().plot(kind='bar')
plt.show()
```
