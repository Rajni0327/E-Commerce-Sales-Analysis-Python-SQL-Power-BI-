```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cleaned_ecommerce_data.csv")
```

```python
platform_revenue = df.groupby('Platform')['Order Value (INR)'].sum()
monthly_sales = df.groupby('Month')['Order Value (INR)'].sum()
```

```python
print(platform_revenue)
```

```python
refund_counts = df['Refund Requested'].value_counts()

print("\nRefund Requested Counts:")
print(refund_counts)
```

```python
refund_by_platform = df[df['Refund Requested'] == 1].groupby('Platform').size()

print("\nRefund Count by Platform:")
print(refund_by_platform)

refund_by_platform = refund_by_platform.sort_values()
refund_by_platform.plot(kind='line', marker='o')

plt.title('Total Refunds by Platform')
plt.xlabel('Platform')
plt.ylabel('Number of Refunds')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
refund_by_platform = refund_by_platform.sort_values()
refund_by_platform.plot(kind='bar')

plt.title('Total Refunds by Platform')
plt.xlabel('Platform')
plt.ylabel('Number of Refunds')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
delay_counts = df['Delivery Delay'].value_counts()

print("Delivery Delay Counts:")
print(delay_counts)
```

# Revenue by platform

```python
# Sum of revenue per platform
revenue_by_platform = df.groupby('Platform')['Order Value (INR)'].sum()

print("\nRevenue by Platform:")
print(revenue_by_platform)



revenue_by_platform = revenue_by_platform.sort_values()
# Plot LINE graph
revenue_by_platform.plot(kind='line', marker='o')

plt.title('Revenue Generated by Each Platform')
plt.xlabel('Platform')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

plt.show()
```
