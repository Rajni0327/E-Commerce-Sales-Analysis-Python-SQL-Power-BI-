```python
import pandas as pd
import numpy as np

df = pd.read_csv("Ecommerce_Delivery_Analytics_New.csv")

# basic info
print(df.shape)
print(df.columns)
print(df.info())
print(df.describe())
```

```python
df.head()
```

# Data Cleaning

```python
# convert datetime
df['Order Date & Time'] = pd.to_datetime(
    df['Order Date & Time'], 
    errors='coerce'
)
```

```python
# drop invalid rows
df = df.dropna(subset=['Order Date & Time'])
```

```python
df.head()
```

```python
# creating time features
df['Order Date'] = df['Order Date & Time'].dt.date
df['Month'] = df['Order Date & Time'].dt.month
df['Hour'] = df['Order Date & Time'].dt.hour
```

```python
df.head()
```

```python
# convert categorical flags to numeric
df['Delivery Delay'] = df['Delivery Delay'].map({'Yes':1, 'No':0})
df['Refund Requested'] = df['Refund Requested'].map({'Yes':1, 'No':0})
```

```python
df.head()
```

```python
delay_counts = df['Delivery Delay'].value_counts()

print("Delivery Delay Counts:")
print(delay_counts)
```

```python
refund_counts = df['Refund Requested'].value_counts()

print("\nRefund Requested Counts:")
print(refund_counts)
```

```python
summary = pd.DataFrame({
    'status': ['Delivery Delay', 'Refund Requested'],
    'Count_1': [
        (df['Delivery Delay'] == 1).sum(),
        (df['Refund Requested'] == 1).sum()
    ],
    'Count_0': [
        (df['Delivery Delay'] == 0).sum(),
        (df['Refund Requested'] == 0).sum()
    ],
    'Percentage (%)': [
        df['Delivery Delay'].mean() * 100,
        df['Refund Requested'].mean() * 100
    ]
})

print(summary)
```

```python
# check missing
print(df.isnull().sum())
```

```python
# drop duplicates
df = df.drop_duplicates()
```

```python
df.head()
```

```python
# revenue metrics
total_revenue = df['Order Value (INR)'].sum()
avg_order_value = df['Order Value (INR)'].mean()

# delivery metrics
avg_delivery_time = df['Delivery Time (Minutes)'].mean()
delay_rate = df['Delivery Delay'].mean() * 100

# refund metrics
refund_rate = df['Refund Requested'].mean() * 100
```

```python
print(total_revenue, avg_order_value, avg_delivery_time, delay_rate, refund_rate)
```

```python
rounded_values = {
    "Total Revenue": round(total_revenue, 2),
    "Avg Order Value": round(avg_order_value, 2),
    "Avg Delivery Time": round(avg_delivery_time, 2),
    "Delay Rate (%)": round(delay_rate, 2),
    "Refund Rate (%)": round(refund_rate, 2)
}

print(rounded_values)
```

# grouped analysis

```python
# revenue by platform
platform_revenue = df.groupby('Platform')['Order Value (INR)'].sum().sort_values(ascending=False)

# revenue by category
category_revenue = df.groupby('Product Category')['Order Value (INR)'].sum().sort_values(ascending=False)

# delay vs rating
delay_rating = df.groupby('Delivery Delay')['Service Rating'].mean()

# refund vs delay
refund_delay = df.groupby('Delivery Delay')['Refund Requested'].mean()

# monthly sales
monthly_sales = df.groupby('Month')['Order Value (INR)'].sum()
```

```python
print(platform_revenue)
print("--------------------------------")
print(category_revenue)
print("--------------------------------")
print(delay_rating) 
print("--------------------------------")
print(refund_delay) 
print("--------------------------------")
print(monthly_sales)
```
